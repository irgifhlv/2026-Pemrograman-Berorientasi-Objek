##PRAKTIKUM 1

```python
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print(f"{self.nama} terbang dengan cara umum.")

    def bersuara(self):
        print(f"{self.nama} mengeluarkan suara burung.")


class Elang(Burung):
    def __init__(self, nama, rentang_sayap):
        super().__init__(nama)
        self.rentang_sayap = rentang_sayap

    def terbang(self):
        print(f"{self.nama} terbang tinggi di angkasa.")

    def bersuara(self):
        print(f"{self.nama} berteriak nyaring!")


class Pipit(Burung):
    def __init__(self, nama, warna):
        super().__init__(nama)
        self.warna = warna

    def terbang(self):
        print(f"{self.nama} terbang cepat di pepohonan.")

    def bersuara(self):
        print(f"{self.nama} berkicau: Cit cit!")


def demo(burung_list):
    for b in burung_list:
        b.terbang()
        b.bersuara()
        print("-"*15)


if __name__ == "__main__":
    data = [
        Elang("Elang Jawa", 1.5),
        Pipit("Pipit", "Coklat"),
        Burung("Burung Biasa")
    ]
    demo(data)
```

##PRAKTIKUM 2

```python
class Kucing:
    def bersuara(self):
        print("Meow")


class Anjing:
    def bersuara(self):
        print("Guk guk")


class Mobil:
    def jalan(self):
        print("Brummm")


def bunyi(obj):
    try:
        obj.bersuara()
    except AttributeError:
        print("Tidak bisa bersuara")


if __name__ == "__main__":
    data = [Kucing(), Anjing(), Mobil()]
    for d in data:
        bunyi(d)
```

##PRAKTIKUM 3

```python
class Kalkulator:
    def tambah(self, *angka):
        total = 0
        for a in angka:
            if isinstance(a, (int, float)):
                total += a
        print("Hasil:", total)


if __name__ == "__main__":
    k = Kalkulator()
    k.tambah(1, 2)
    k.tambah(1, 2, 3, 4)
```

##PRAKTIKUM 5

```python
import math

class Bentuk:
    def hitung_luas(self):
        raise NotImplementedError


class Lingkaran(Bentuk):
    def __init__(self, r):
        self.r = r

    def hitung_luas(self):
        return math.pi * self.r ** 2


class Persegi(Bentuk):
    def __init__(self, s):
        self.s = s

    def hitung_luas(self):
        return self.s * self.s


class Taplak:
    def __init__(self, p, l):
        self.p = p
        self.l = l

    def hitung_luas(self):
        return self.p * self.l


def tampil(obj):
    try:
        print("Luas:", obj.hitung_luas())
    except:
        print("Tidak bisa dihitung")


if __name__ == "__main__":
    data = [Lingkaran(7), Persegi(5), Taplak(2, 3)]
    for d in data:
        tampil(d)
```

##PRAKTIKUM 6

```python
class Burung:
    def __init__(self, nama):
        self.nama = nama

    def terbang(self):
        print("Terbang")


class Elang(Burung):
    def berburu(self):
        print("Berburu")


def cek(obj):
    if isinstance(obj, Elang):
        obj.berburu()
    elif isinstance(obj, Burung):
        obj.terbang()
    else:
        print("Bukan burung")


if __name__ == "__main__":
    cek(Elang("Elang"))
    cek(Burung("Merpati"))
    cek("String")
```

##PENUGASAN

```python
class Komputer:
    def __init__(self, merk, processor, ram):
        self.merk = merk
        self.processor = processor
        self.ram = ram

    def info_spesifikasi(self):
        print(f"Merk: {self.merk}")
        print(f"Processor: {self.processor}")
        print(f"RAM: {self.ram} GB")

    def jalankan_aplikasi(self, app):
        print(f"{self.merk} menjalankan {app}")

    # overloading pakai *args
    def install_software(self, *apps):
        for a in apps:
            print(f"Install {a} di {self.merk}")


class Laptop(Komputer):
    def __init__(self, merk, processor, ram, layar, berat):
        super().__init__(merk, processor, ram)
        self.layar = layar
        self.berat = berat

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f"Layar: {self.layar} inch")
        print(f"Berat: {self.berat} kg")


class Desktop(Komputer):
    def __init__(self, merk, processor, ram, casing, monitor):
        super().__init__(merk, processor, ram)
        self.casing = casing
        self.monitor = monitor

    def info_spesifikasi(self):
        super().info_spesifikasi()
        print(f"Casing: {self.casing}")
        print(f"Monitor: {'Ya' if self.monitor else 'Tidak'}")


def cetak(data):
    for d in data:
        print("-"*20)
        d.info_spesifikasi()


if __name__ == "__main__":
    l = Laptop("Asus", "Ryzen 5", 16, 14, 1.5)
    d = Desktop("HP", "i7", 32, "Tower", True)

    daftar = [l, d]
    cetak(daftar)

    print("\nInstall Software:")
    l.install_software("Python", "VS Code")
```
